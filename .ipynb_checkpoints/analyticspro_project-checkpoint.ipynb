```python
def get_fmi_data (fmi_id, start_date, end_date):
    url = 'http://data.fmi.fi/fmi-apikey/'+credentials.login['fmi_api_key']+'/wfs?request=getFeature&storedquery_id=fmi::observations::weather::daily::multipointcoverage&fmisid='+fmi_id+'&starttime='+start_date+'&endtime='+end_date
    return requests.get(url)
```

```python
fmi_id = '101004'
start_date = '2009-01-01'
end_date = '2009-12-31'
r = get_fmi_data(fmi_id, start_date, end_date)
```

```python
root = ElementTree.fromstring(r.content)
```

```python
# print all tags in xml file
for child in root.iter('*'):
    print(child.tag)
```

```python
for item in root.iter('{http://www.opengis.net/swe/2.0}DataRecord'):
    print(item)
    for record in item.iter():
        print(record.findtext('name'))
```

```python
for position in root.iter('{http://www.opengis.net/gmlcov/1.0}positions'):
    position_text = position.text
    pos_list = []
    for row in position_text.split('\n')[1:-1]:
        pos_list.append(row.split())
        date_list = []
        for row in pos_list:
            date_list.append(datetime.fromtimestamp(int(row[2])).strftime('%Y-%m-%d'))
```

```python
date_list
```

```python
for observation in root.iter('{http://www.opengis.net/gml/3.2}doubleOrNilReasonTupleList'):
    observation_text = observation.text
    obs_list = []
    for row in observation_text.split('\n')[1:-1]:
        obs_list.append(row.split())
```
